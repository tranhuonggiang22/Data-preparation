```python
import pandas as pd
```

```python
#Ex1
dt_ap = pd.read_csv('/content/drive/MyDrive/DATA PREPERATION/buoi 4/aapl.csv', index_col = [0], parse_dates = [0])
dt_am = pd.read_csv('/content/drive/MyDrive/DATA PREPERATION/buoi 4/amzn.csv', index_col = [0], parse_dates = [0])
dt_fb = pd.read_csv('/content/drive/MyDrive/DATA PREPERATION/buoi 4/fb.csv', index_col = [0], parse_dates = [0])
dt_go = pd.read_csv('/content/drive/MyDrive/DATA PREPERATION/buoi 4/goog.csv', index_col = [0], parse_dates = [0])
dt_nf = pd.read_csv('/content/drive/MyDrive/DATA PREPERATION/buoi 4/nflx.csv', index_col = [0], parse_dates = [0])
```

```python
#a
dt_ap['ticker'] = 'APPL'
dt_am['ticker'] = 'AMZ'
dt_fb['ticker'] =  'FB'
dt_go['ticker'] = 'GOOG'
dt_nf['ticker'] = 'NETF'

faang = pd.concat([dt_ap,dt_am,dt_fb,dt_go,dt_nf], axis = 0, join = 'inner').sort_index()
faang
```

```python
#b
faang.to_csv('/content/drive/MyDrive/DATA PREPERATION/buoi 4/faang.csv')
faang
```

```python
#c
faang.index = pd.to_datetime(faang.index)
faang
```
