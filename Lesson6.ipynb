```python
fb = pd.read_csv("/content/drive/MyDrive/DATA PREPERATION/buoi6/fb_stock_prices_2018.csv")
fb.head(5)
```

```python
#EX2 by day
y = fb.high - fb.low
x = fb.date
plt.plot(x,y)
plt.xlabel("day")
plt.ylabel("MEDIUM(X)")
```

```python
x = fb.assign(week = pd.DatetimeIndex(fb.date).week)
x.head()
```

```python
dif = x.groupby(['week']).agg({"high":max, "low": min})
dif.head(5)
```

```python
dif['diffirence'] = dif['high']-dif['low']
```

```python
#EX2 by week
plt.plot(dif['diffirence'])
plt.xlabel("week")
plt.ylabel("MEDIUM(X)")
```
